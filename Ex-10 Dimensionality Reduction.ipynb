```python
from sklearn import datasets
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
df = pd.DataFrame(iris['data'], columns=iris['feature_names'])
```

```python
print("Original Data:")
print(df.head())
```

```python
scaler = StandardScaler()
scaled_data = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
print("\nScaled Data:")
print(scaled_data.head())
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(scaled_data.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Scaled Data')
plt.show()
```

```python
pca = PCA(n_components=3)
pca.fit(scaled_data)
data_pca = pca.transform(scaled_data)
data_pca = pd.DataFrame(data_pca, columns=['PC1', 'PC2', 'PC3'])
```

```python
print("\nPCA Transformed Data:")
print(data_pca.head())
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data_pca.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of PCA Data')
plt.show()
```
